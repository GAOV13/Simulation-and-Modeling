La incertidumbre económica, frecuentemente medida a través del índice de incertidumbre de política económica (EPU, por sus siglas en inglés), es un concepto central para comprender las dinámicas económicas y financieras. Los altos niveles de incertidumbre suelen asociarse con impactos negativos en el consumo, la inversión y el crecimiento económico, especialmente en regiones vulnerables como América Latina​ [1] [2].

**América Latina y la Incertidumbre Económica**

Debido a sus características estructurales, América Latina enfrenta desafíos particulares frente a la incertidumbre:

1. **Dependencia de materias primas:** Los choques en los precios de las materias primas, como el petróleo y el carbón, afectan tanto la volatilidad de la tasa de cambio como la estabilidad macroeconómica​ [3].
2. **Impacto del COVID-19:** La pandemia aumentó la incertidumbre económica, con caídas significativas en la actividad económica y aumentos en la volatilidad del mercado​ [2].



**Colombia: Un Caso Relevante**

**Estudios de la incertidumbre economica en otros mercados (China)**

El artículo "Economic policy uncertainty and stock market volatility in China: Evidence from SV-MIDAS-t model" explora cómo la incertidumbre de política económica (EPU) afecta la volatilidad del mercado accionario en China. Se utiliza un enfoque innovador que combina un modelo de volatilidad estocástica (SV) con una estructura de muestreo de datos mixtos (MIDAS) bajo una distribución t. Este modelo permite capturar características clave de los datos financieros, como colas pesadas, exceso de curtosis y autocorrelación débil en la volatilidad​ [4].

Este hallazgo subraya la importancia del EPU como predictor de riesgos financieros y su capacidad para mejorar la precisión de los modelos de volatilidad​ [4].



## ¿Qué es el índice EPU y cómo se calcula?

El índice de incertidumbre de política económica (EPU, por sus siglas en inglés) mide el nivel de incertidumbre derivado de cambios en las políticas económicas de un país. Este índice refleja el impacto que las decisiones y variaciones en las políticas gubernamentales pueden tener sobre las expectativas de los agentes económicos, como inversionistas y consumidores.






**Cálculo del índice EPU**

El cálculo del índice EPU se basa en un enfoque metodológico que combina análisis de contenido en medios de comunicación y técnicas estadísticas. Los pasos principales son los siguientes:

1. **Análisis de contenido en medios**: Se realiza un análisis textual en los medios de comunicación, principalmente artículos periodísticos, para identificar palabras clave relacionadas con "incertidumbre", "economía", y "política". Esta es una forma de capturar cómo se discuten los cambios políticos y económicos en la prensa, lo cual refleja la percepción de incertidumbre en la sociedad.

2. **Frecuencia relativa:** Se cuenta la frecuencia de los artículos que mencionan estas palabras clave. La frecuencia se calcula como la proporción de artículos con las palabras clave frente al total de publicaciones, lo que da una medida relativa de la cobertura de la incertidumbre económica en los medios.

3. **Ajuste de escala y diferencias logarítmicas:** Para garantizar la consistencia temporal del índice, se ajusta la escala mediante diferencias logarítmicas entre los valores mensuales del índice. Este ajuste es crucial, ya que permite captar los cambios relativos en el índice, evitando que las fluctuaciones sean interpretadas como cambios absolutos que podrían no reflejar la magnitud real de la incertidumbre. Además, el ajuste ayuda a que los datos sean estacionarios, lo cual es importante para el análisis económico.

**Importancia del EPU para la volatilidad del mercado**

El índice EPU tiene una relación directa con la volatilidad de los mercados financieros. A continuación, se detallan algunos de sus impactos más relevantes encontrados en [4]:

* **Impacto directo:** Incrementos en el EPU afectan las expectativas de los inversionistas, quienes ajustan sus decisiones ante la incertidumbre. Esto genera un aumento en la volatilidad de los mercados financieros, ya que los agentes económicos tienden a reaccionar de forma más impredecible cuando perciben que las políticas económicas pueden cambiar de manera inesperada.

* **Efecto de largo plazo:** El EPU no solo influye a corto plazo, sino que también tiene efectos sobre la volatilidad a largo plazo. Los impactos de un aumento en el EPU pueden observarse meses después del cambio inicial, especialmente en los mercados bursátiles y otros activos financieros. Esto ocurre porque las expectativas de los inversionistas se ajustan gradualmente a medida que los efectos de los cambios políticos se hacen más evidentes.

* **Mejora en modelos predictivos:** Integrar el EPU en modelos predictivos de volatilidad, como el modelo SV-MIDAS-t, mejora la precisión de las predicciones a largo plazo. Estos modelos pueden capturar mejor las dinámicas del mercado, especialmente en contextos donde los cambios políticos y económicos son fuente de incertidumbre. La inclusión del EPU permite incorporar los efectos de esta incertidumbre en las proyecciones de volatilidad, lo que mejora los modelos de riesgo financiero.

### Imports

```python
import numpy as np
from scipy.stats import t
from scipy.optimize import minimize
import pandas as pd
import matplotlib.pyplot as plt
```

### Datos del mercado Colombiano

```python
excel = './content/TRM_20A.xlsx'
dfTRM = pd.read_excel(excel, sheet_name='Sheet1')
excel = "./content/Colombia_Policy_Uncertainty_Data_GS.xlsx"
dfEPU = pd.read_excel(excel, sheet_name='Year')

TRM = dfTRM['TRM'].to_numpy()
TRM = np.array([0]+[np.log(TRM[i+1])-np.log(TRM[i]) for i in range(len(TRM)-1)])
fecha_TRM = dfTRM['Fecha'].to_numpy()
EPU = dfEPU['EPUC'].to_numpy()
EPU = np.array([0]+[np.log(EPU[i+1])-np.log(EPU[i]) for i in range(len(EPU)-1)])
fecha_EPU = dfEPU['Date'].to_numpy()
```

### Modelos

```python
def midas_weights(w1, w2, K=36):
    k = np.arange(1, K+1)
    weights = (k/K)**(w1-1) * (1 - k/K)**(w2-1)
    return weights / np.sum(weights)

def long_term_component(EPU, theta, m, w1, w2):
    K = len(EPU)
    weights = midas_weights(w1, w2, K)
    ln_tau_t = m + theta * np.dot(weights, EPU[::-1])
    tau_t = np.exp(ln_tau_t)
    return tau_t

def calculate_volatility(TRM_daily, EPU, phi, gamma, theta, m, w1, w2, v):
    K = len(EPU)
    tau_t = long_term_component(EPU, theta, m, w1, w2)
    h = np.zeros(len(TRM_daily))
    eps_t = np.zeros(len(TRM_daily))

    for i in range(1, len(TRM_daily)):
        h[i] = phi * h[i-1] + gamma * t.rvs(df=v)
        eps_t[i] = TRM_daily[i] - np.exp(0.5 * h[i] + 0.5 * tau_t)

    estimated_volatility = np.exp(0.5 * h + 0.5 * tau_t)
    return estimated_volatility

def garch_midas_t_errors(data, EPU, alpha, beta, theta, m, w1, w2, omega, v):
    """
    GARCH-MIDAS con errores distribuidos como t-Student.

    Parámetros:
    - data: Rendimientos diarios (serie temporal).
    - EPU: Índice de incertidumbre económica.
    - alpha, beta: Parámetros GARCH.
    - theta, m: Parámetros para la componente de largo plazo.
    - w1, w2: Pesos MIDAS.
    - omega: Constante de varianza GARCH.
    - v: Grados de libertad de la distribución t-Student.
    """
    # Componente a largo plazo
    tau_t = long_term_component(EPU, theta, m, w1, w2)
    sigma_t = np.zeros(len(data))
    sigma_t[0] = np.sqrt(omega)  # Inicialización

    # Simulación de errores t-Student
    errors = np.random.standard_t(v, size=len(data))

    for t in range(1, len(data)):
        residual = data[t - 1] - errors[t - 1] * sigma_t[t - 1] if t > 0 else 0
        sigma_t[t] = np.sqrt(omega + alpha * residual**2 + beta * sigma_t[t - 1]**2)

    # Varianza condicional final
    conditional_volatility = sigma_t * np.sqrt(tau_t)
    return conditional_volatility
```

### Gráfico

```python
# Calcular la volatilidad histórica con ventana móvil de 20 días
window_size = 30
historical_volatility = np.array([np.std(TRM[i-window_size:i]) for i in range(window_size, len(TRM))])
historical_volatility = np.concatenate((np.zeros(window_size), historical_volatility))
mean_hv = np.mean(historical_volatility)

# Calcular la volatilidad estimada usando el modelo SV-MIDAS-t
# Parametros:
# Parámetros iniciales
phi = 0.9961
gamma = 0.1003
theta = 3.9657
m = -8.6654
w1, w2 = 2.0068, 5.9494
v = 6.5171

estimated_volatility = (calculate_volatility(TRM, EPU, phi, gamma, theta, m, w1, w2, v))
mean_ev = np.mean(estimated_volatility)
fact = mean_hv/mean_ev
estimated_volatility = fact * estimated_volatility

# Calcular la volatilidad estimada usando el modelo Garch-MIDAS-t
# Parametros:
# Parámetros iniciales
omega = 1.3e-7
alpha = 0.0439
beta = 0.9141
m = -10.0976
theta = 2.1602
w1, w2 = 1.8466, 6.3224
v = 5.2190

estimated_garch = (garch_midas_t_errors(TRM, EPU, alpha, beta, theta, m, w1, w2, omega, v))
mean_ev = np.mean(estimated_garch)
fact = mean_hv/mean_ev
estimated_garch = fact * estimated_garch

plt.plot(fecha_TRM, estimated_volatility, color='blue', label='Volatilidad Condicional (sv-midas)')
plt.plot(fecha_TRM, historical_volatility[:len(historical_volatility)], "r*", label="Volatilidad Estimada Historica", markersize=1.5)
plt.plot(fecha_TRM, estimated_garch, color='green', label='Volatilidad Condicional (Garch-midas)')
plt.legend()
plt.grid()
plt.title("Volatilidad")
plt.show()
```

```python
# Fechas de análisis
inicio_pandemia = np.datetime64('2020-01-01')
fin_pandemia = np.datetime64('2022-12-31')
idx_pandemia = (fecha_TRM >= inicio_pandemia) & (fecha_TRM <= fin_pandemia)


# Gráfico de la pandemia
plt.plot(fecha_TRM[idx_pandemia], historical_volatility[idx_pandemia], "r*", label="Volatilidad Histórica", markersize=1.5)
plt.plot(fecha_TRM[idx_pandemia], estimated_volatility[idx_pandemia], label="Volatilidad SV-MIDAS", color="blue")
plt.plot(fecha_TRM[idx_pandemia], estimated_garch[idx_pandemia], label="Volatilidad GARCH-MIDAS", color="green")
plt.title("Volatilidad Durante la Pandemia")
plt.legend()
plt.grid()
plt.show()
```
